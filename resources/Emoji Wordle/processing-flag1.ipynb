```python
filename = 'all-emojis.txt'
lst = []
with open(filename) as f:
    lst = f.readlines()
lst = [l for l in lst if not (l.startswith('#') or l.startswith('\n') or l.startswith('https'))]
lst = [l.split() for l in lst]
lst = [l[4] for l in lst if l[1] == ';']
group_n = len(lst) // 64 + 1
for i in range(group_n):
    print(''.join(lst[i*64:(i+1)*64]))
```
